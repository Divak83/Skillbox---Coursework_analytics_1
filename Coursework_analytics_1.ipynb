### Для начала сохраним данные из файлов в датасеты

```python
# основной датасет
df = pd.read_csv('courses.csv')
df1 = pd.read_csv('course_contents.csv')
df2 = pd.read_csv('progresses.csv')
df3 = pd.read_csv('progress_phases.csv')
# для корректного объединения датасетов необходимо часть столбцов переименовать
df.columns = ['course_id', 'course_title', 'course_field']
df2.columns = ['progress_id', 'student_id', 'course_id']
# объединим все датасеты в один
main= df3.merge(df2)
main = main.merge(df)
main = main.merge(df1)
# выведем первые 5 строк датасета
main.head()
```

### Далее узнаем количество студентов на каждом курсе. Посчитаем минимальный, максимальный, средний, медианный возраст для всех студентов и для каждого курса отдельно.

```python
# Посчитаем количествo учеников на каждом курсе
stud_count = main.groupby(['course_title'], as_index = False)['student_id']\
    .nunique().rename(columns={'student_id': 'num_stud'}).sort_values('num_stud')
stud_count
```

```python
# Посчитаем минимальный, максимальный, средний, медианный возраст студентов

# При проверке были обнаружены ошибки в данных. Были указаны невозможные даты рождения.
students['birthday'].apply(['min', 'max'])
```

```python
# Укажем минимальный возраст студента в 15 лет, а максимальный в 70 лет.
students['age'] = (pd.to_datetime('today') - students['birthday']).astype('<m8[Y]')
students = students[(students['age'] >= 15.0) & (students['age'] <= 70.0)]
students['age'].agg(['min', 'max', 'mean', 'median'])
```

```python
# Посчитаем минимальный, максимальный, средний, медианный возраст студентов на каждом курсе
# для расчета создадим отдельный датасет объединив датасеты main и students
stud_age = pd.merge(main, students, on = 'student_id', how ='left')
stud_age.groupby('course_title')['age'].agg(['min', 'max', 'mean', 'median'])
```

### Построим график отображающий количество студентов на кажом курсе.

```python
#Построим bar-chart, отражающий количество студентов на каждом курсе.Ticks развернем так, чтобы они были читаемы

# зададим формат и нарисуем рамку графика
fig = plt.figure(figsize=(12,8))
ax= plt.subplot(111)

# посторим bar-chart по полученным данным
ax.bar(stud_count.course_title, stud_count.num_stud) 

# развернем подписи курсов, чтоб они были читаемы
plt.xticks(rotation=90) 

# добавим сетку
ax.yaxis.grid(linestyle='-', linewidth=2) 
ax.set_axisbelow(True) 

# добавим заголовок
ax.set_title('Kоличество студентов на каждом курсе', fontsize=14)  
plt.show
```

### Дополним информацию на графике выделив разными цветами сферы деятельности, к которым относятся курсы. Также укажем медианное количество студентов на курсах.

```python
# Постройте горизонтальный (столбцы должны располагаться горизонтально) bar-chart, отражающий количество студентов 
# на каждом курсе. График должен иметь заголовок. Значения должны быть отсортированы.
# Цвет столбцов должен содержать информацию о сфере, к которой относится курс (то есть нужна легенда).
# Прозрачность должна стоять на отметке 0.1. На график должна быть нанесена линия медианы. 
# У медианы должен быть свой цвет. Рамки у графика быть не должно ⭐

# сгруппируем данные по названию курса и сфере деятельности, подсчтитаем количество студентов на каждом курсе.
stud_num =main.pivot_table(values='student_id', index='course_title', columns='course_field',\
                        aggfunc=lambda x: x.unique().size, margins=True)\
    .sort_values('All',ascending=True).drop(index=['All'], columns=['All'])

# создадим график и зададим интенсивность цвета (яркость) столбцов 0.1 и pазмер графика
fig, ax = plt.subplots(figsize = (15,10))
ax = stud_num.plot(kind='barh', ax = ax, stacked = True, alpha = 0.1) 

# добавляем легенду и форматируем её
plt.legend(loc = 'lower right', title = 'Сфера деятельности', shadow = True, fontsize = 12)

# найдем медиану и добавим ее на  график, зададим вид, ширину, цвет линии 
median = stud_count.num_stud.median()
plt.axvline(x = median, linestyle = '--', linewidth = 1, color = 'black')

# убираем границы графика и добавим заголовок
plt.box(on=None) 
ax.set_title('Kоличество студентов на каждом курсе', fontsize=14)
plt.show()
```

### Рассчитаем прирост студентов на каждом курсе в каждом месяце за всю историю. Будем считать дату начала прохождения курса студентом по дате начала первой домашней работы.

```python
# Узнаем в каком модуле каждого курса, встречается первое домашнее задание
main[(main['is_homework'] == True)].groupby('course_title').min()
```

```python
# В курсе 'Веб-дизайн Базовый' первое домашее задание в 3 модуле, в остальных в 1 модуле.

# Создадим новый датасет, где будут только строки с первым домашним заданием.
growth = main[(main['is_homework'] == True) & (main['module_number'] == 1) \
                    | ((main['module_number'] == 3) & (main['course_title'] == 'Веб-дизайн Базовый'))]
growth.head()
```

```python
# перенесем колоноку 'start_date' в индекс
growth.index = growth.start_date
# сгруппируем данные по названию курса и месяцам
growth = growth.groupby('course_title').resample('M').student_id.nunique().unstack(0).fillna(0).astype('int')
# в датах оставим только год и месяц
growth.index = pd.to_datetime(growth.index).strftime('%Y-%m')
growth.head()
```

### На основании полученных данных постром графики с приростом студентов в каждом месяце. Каждый график будет иметь заголовок .

#### Расчитаем  конверсию перехода студентов из одного модуля в другой на каждом курсе.

```python
# Для расчета будем использовать следующую формулу: 
# отношение количества студентов, приступивших к выполнению домашнего задания в этом модуле
# (если дз в модуле несколько, то считать по первому дз в модуле),
# к количеству студентов, сдавших задание в предыдущем модуле (если дз в модуле несколько, то считать по последнему дз в модуле).

# создадим новый датасет
convers = main[main['is_homework'] == True][['student_id', 'module_number', 'start_date', \
                                             'finish_date', 'course_title', 'module_title', 'course_field']]

# сгрупируем данные оставив первую дату начала д/з и последную сдачи д/з
convers = convers.groupby(['course_title', 'module_number', 'module_title','student_id', 'course_field'])\
                                .agg({'start_date': 'min', 'finish_date': 'max'}).reset_index()
convers.head()

```

```python
# посчитаем количество начатых и сданных д/з
convers = convers.groupby(['course_title', 'module_number', 'module_title', 'course_field'])\
                [['start_date', 'finish_date']].count().reset_index()\
                .rename(columns={'start_date':'started', 'finish_date':'finished'})

# свдиг столбца finished вниз в рамках конкретного курса
convers['finished'] = convers.groupby('course_title')['finished'].shift()

# вычислим конверсию 
convers['conv'] = (convers['started'] / convers['finished']).fillna(1)

# создадим столбец с конверсией в процентах
convers['conv_perc'] = convers['conv'] * 100
convers.head()
```

### Построим графики отражающие конверсию перехода студентов из одного модуля в другой. 

```python
# Построим 15 графиков. Добавим заголовки и подпишем оси. 
# Ticks развернем так, чтобы они были читаемы.

# создадим список с названиями курсов
course_list = list(convers.course_title.unique())

# по очереди строим каждый график
for i in range(15):
    # сохраняем список номеров модулей и значения конверсии для текущего курса
    modules_numbers_list = list(convers[convers['course_title'] == course_list[i]].module_number.astype('str'))
    conversions = convers[convers['course_title'] == course_list[i]]['conv_perc'].values
    
    # зададим размер и построим график
    plt.figure(figsize=(15, 5))   
    plt.bar(modules_numbers_list, conversions, alpha=0.8)
    
    # название графика
    plt.title(course_list[i], fontsize=20)
    
    # зададим параметры оси х и у
    plt.ylabel('Конверсия, %', fontsize=16)
    plt.xlabel('Номер модуля', fontsize=14)
    
    # добавим надписи
    plt.xticks(modules_numbers_list)
    
    plt.show();
```


### Дополним информацию на графике выделив разными цветами сферы деятельности, к которым относятся курсы. Добавим  линию медианы конверсии для каждого курса. 

```python
# Построим 15 горизонтальныx bar-chart, отражающий конверсию перехода студентов из одного модуля в другой на каждом курсе.
# 15 графиков. Графики будут иметь заголовки.
# Ticks будут  содержать номер и название модуля. Добавим легенда о сфере деятельности. Прозрачность выставим  на отметке 0.1.
# На графиках должна быть нанесена. У медианы будет свой цвет. Рамок у графиков не будет ⭐


# удалим ненужные столбцы
convers = convers.drop(columns=[ 'conv', 'started', 'finished'])
convers.head()

colors = {'Marketing': 'yellow',
 'Design': 'red',
 'Development': 'blue',
 'Business': 'green'}

# добавим столбец с цветами для каждой сферы
convers['colors'] = convers['course_field'].map(colors)

# отсортируем по номеру модуля в обратном порядке
convers = convers.sort_values(by=['course_title', 'module_number'], ascending=False)
convers.head()    
```

```python
# по очереди построим каждый график
for i in range(15):

    # создадим список модулей и номеров модулей курса
    modules = list(convers[convers['course_title'] == course_list[i]]['module_title'])
    numbers_list = convers[convers['course_title'] == course_list[i]]['module_number']
    
    # сохраним в переменную значения конверсии модулей курса
    conversions = convers[convers['course_title'] == course_list[i]]['conv_perc'].values
    
    # вычислим медиану 
    median = int(np.median(conversions)) 
        
    # зададим цвет и название сферы курса
    color = convers[convers['course_title'] == course_list[i]]['colors'].values
    field = convers[convers['course_title'] == course_list[i]]['course_field'].values[0]    
    
    # список с нумерацией и названием модулей текущего курса
    mod_name = [str(number) + '. ' + title for number, title in zip(numbers_list, modules)]
    
    # построим график
    plt.barh(mod_name, conversions, label = course_list[i], color = color, alpha = 0.1)
    
    # зададим заголовок графика
    plt.title(course_list[i], fontsize=20)
    
    # добавим медиану на график
    plt.axvline(median, color='k', linestyle='--', label='медиана')
    
    # добавим названия модулей на график
    plt.yticks(mod_name)
    
    # зададим пареметры оси x
    plt.xlabel('Конверсия, %', fontsize=16)    
    
    # добавим легенду и уберем рамки у графика
    plt.legend(labels = ['медиана', field],  bbox_to_anchor=(1,1))
    plt.box(on=None)
    
    plt.show()
```
